```python
# you can us argument marker to emphasize each point with specified marker.

import matplotlib.pyplot as plt
import numpy as np
ypoints = np.array([3,8,1,10])
plt.plot(ypoints,marker = 'o')
plt.show()
```

# format string " fmt" -
marker|line|color

```python

import matplotlib.pyplot as plt
import numpy as np

ypoints=np.array([3,8,1,2,0])
plt.plot(ypoints,'o:r')
plt.show()

# line reference
#  - solid line
# : dotted line
# -- dashed line
# -. dashline/dotted line

# color reference
# r - red
# g - green
# b - blue
# c - cyan
# m - magenta
# y - yellow
# k - black
# w - white
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([3,2,5,7,19])
plt.plot(y,linestyle="dotted")
plt.show()
```

```python
# Marker Size
import matplotlib.pyplot as plt
import numpy as np
y= np.array([1,4,27,4,8,19])
plt.plot(y , marker = 'o', ms = 25)
plt.show()
```

```python
# marker color - mec
import matplotlib.pyplot as plt
import numpy as np
y=np.array([2,4,7,18,3])
plt.plot(y, marker = 'o', ms=20, mec = 'r')
plt.show()
```

```python
# marker face color - mfc
import matplotlib.pyplot as plt
import numpy as np
y=np.array([3,2,5,7,19])
plt.plot(y,marker = 'o',  ms = 20, mfc='r')
plt.show()
```
