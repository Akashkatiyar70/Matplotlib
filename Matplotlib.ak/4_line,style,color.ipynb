# linestyle or ls
is used to changed the style of the plotted line

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([3,2,5,7,19])
plt.plot(y,linestyle = 'dotted')
plt.show()
```

# line color - c

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([3,8,1,10])
plt.plot(y,color='r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y=np.array([2,8,5,7,19,20,32])
plt.plot(y,linewidth="20",color='m',linestyle="dotted")
plt.show()
```

```python
# example line

import matplotlib.pyplot as plt
import numpy as np
x=np.array([2,5,1,8,27])
y=np.array([3,2,5,7,19])
plt.plot(y,linewidth='20.5')
plt.plot(x,linewidth='20.5')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
