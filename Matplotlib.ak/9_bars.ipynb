```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['A','B','C','D'])
y = np.array([3,8,10,3])
plt.barh(x,y,color="hotpink")
plt.show()
```

# Bar Width
The bar() takes the keyword argument width to set the width of the bars:

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['A','B','C','D'])
y = np.array([3,8,10,3])
plt.bar(x,y,color="hotpink",width=0.2)
plt.show()
```

# Bar Height
The barh() takes the keyword argument height to set the height of the bars:

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(['A','B','C','D'])
y = np.array([3,8,10,3])
plt.barh(x,y,color="hotpink",height=0.2)
plt.show()
```
